## 3. Strength Parameter

Controls how different output is from input:
- 0.0 = identical
- 0.5 = balanced
- 1.0 = complete remake

```python
strengths = {'0.2': 'Subtle', '0.5': 'Balanced', '0.8': 'Strong'}
for s, desc in strengths.items():
    print(f'Strength {s}: {desc}')
```
